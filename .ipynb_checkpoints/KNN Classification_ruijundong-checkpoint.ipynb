```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

### (a) Download the Vertebral Column Data Set from: https://archive.ics.uci.edu/ml/datasets/Vertebral+Column.

```python
df = pd.read_csv("./vertebral_column_data/column_2C.dat", header=None, delimiter = ' ',
                 names = ["pelvic_incidence", "pelvic_tilt", "lumbar_lordosis_angle", "sacral_slope", "pelvic_radius", "spondylolisthesis_grade", "label"])
```

```python
df['label'].replace(['AB','NO'],[1,0],inplace=True)
```

```python
df.head()
```

### (b) Pre-Processing and Exploratory data analysis:

#### i. Make scatterplots of the independent variables in the dataset. Use color to show Classes 0 and 1.

```python
sns.pairplot(df, hue="label",plot_kws={"s": 17}, vars=["pelvic_incidence", "pelvic_tilt", "lumbar_lordosis_angle", "sacral_slope", "pelvic_radius", "spondylolisthesis_grade"])
plt.show()
```

#### ii. Make boxplots for each of the independent variables. Use color to show Classes 0 and 1 (see ISLR p. 129).

```python
f, axes = plt.subplots(3, 2, figsize=(15,12))
y = 0
for col in df.columns[:-1]:
    i, j = divmod(y, 2)
    
    sns.boxplot(x=col, y="label", data=df,orient="h",ax=axes[i, j])
    y=y+1
plt.show()  
```

#### iii. Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set and the rest of the data as the test set.

```python
train0 = df.loc[df['label']==0].iloc[0:70]
train1 = df.loc[df['label']==1].iloc[0:140]

test1 = df.loc[df['label']==1].iloc[140:]
test0 = df.loc[df['label']==0].iloc[70:]
```

### (c) Classification using KNN on Vertebral Column Data Set


#### i. Write code for k-nearest neighbors with Euclidean metric (or use a software package).

```python
train = pd.concat([train0, train1])
test = pd.concat([test0, test1])
X_train = train.iloc[:,0:6]
y_train = train.iloc[:,6]
X_test = test.iloc[:,0:6]
y_test = test.iloc[:,6]
```

#### ii. Test all the data in the test database with k nearest neighbors. 

```python
train_opt = {}
```

```python
K = np.arange(208, 0, -3)
train_err = []
test_err = []

for k in K:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    train_pred = knn.predict(X_train)
    test_pred = knn.predict(X_test)
    
    train_acc = accuracy_score(y_train,train_pred)
    test_acc = accuracy_score(y_test,test_pred)
    
    train_err.append(1-train_acc)
    test_err.append(1-test_acc)
train_opt["(c)ii."] = min(train_err[:-1])
```

```python
#Which k is the most suitable k among those values?
print(K[np.argmin(test_err)], min(train_err[:-1]))
```

```python
plt.plot(K, train_err, label='Train')
plt.plot(K, test_err, label='Test')
plt.xlim(reversed(plt.xlim()))
plt.xlabel('k')
plt.ylabel('Error Rate')
plt.legend(loc='lower left')
plt.show()
```

```python
#Calculate the confusion matrix, true positive rate, true negative rate, precision, and F1-score when k = k.2
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
tpr = tp / (tp + fn)
tnr = tn / (tn + fp)
pcs = tp / (tp + fp)
F = 2 * (pcs * tpr / (pcs + tpr))
print("Confusion Matrix:\n{}\nTrue positive rate: {}\nTrue negative rate: {}\nPrecision: {}\nF_score: {}".format(cm, tpr, tnr, pcs, F))
```

#### iii. Plot the best test error rate,3 which is obtained by some value of k, against the size of training set, when the size of training set is N  {10,20,30,...,210}.4 

```python
N = np.arange(10, 211, 10)
test_err_opt = []
train_err_opt= []

for n in N:
    class0 = math.floor(n/3)
    class1 = n-class0
    train_subset = pd.concat([train0.iloc[0:class0,:],train1.iloc[0:class1,:]])
    
    X_train = train_subset.iloc[:,0:6]
    y_train = train_subset.iloc[:,6]
    
    K = np.arange(1, n+1, 5)
    train_err=[]
    test_err = []
 
    for k in K:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)

        train_pred = knn.predict(X_train)
        test_pred = knn.predict(X_test)

        train_acc = accuracy_score(y_train,train_pred)
        test_acc = accuracy_score(y_test,test_pred)

        train_err.append(1-train_acc)
        test_err.append(1-test_acc)

    opt = K[np.argmin(test_err)]
    test_err_opt.append(min(test_err))
    train_err_opt.append(min(train_err[1:]))
    
train_opt["(c)iii."] = min(train_err_opt)
```

```python
plt.plot(N, test_err_opt)

plt.xlabel('N')
plt.ylabel('Best Rest Error Rate')

plt.show()
```

### (d) Replace the Euclidean metric with the following metrics5 and test them. 

#### i. Minkowski Distance:

```python
#A. which becomes Manhattan Distance with p = 1.
K = np.arange(1, 197, 5)
train_err = []
test_err = []

for k in K:
    knn = KNeighborsClassifier(n_neighbors=k, p=1)
    knn.fit(X_train, y_train)
    
    train_pred = knn.predict(X_train)
    test_pred = knn.predict(X_test)
    
    train_acc = accuracy_score(y_train,train_pred)
    test_acc = accuracy_score(y_test,test_pred)
    
    train_err.append(1-train_acc)
    test_err.append(1-test_acc)
    
train_opt["(d)i.A."] = min(train_err[1:])   
pd.DataFrame(list(zip(K,test_err)), columns = ['K', 'Test Error'])
```

```python
#select the best k when test error is minimum
test_err = np.array(test_err)
K = np.array(K)
K = K[np.where(test_err == test_err.min())]
K
```

```python
#B. with log10(p)  {0.1,0.2,0.3,...,1}. In this case, use the k you found for the Manhattan distance in ??. 
#What is the best log10(p)?

P = 10**np.arange(0.1,1.1,0.1)
train_err = {}
test_err = {}

for k in K:
    test_err[k] = []
    train_err[k] = []
    
    for p in P:
        
        knn = KNeighborsClassifier(n_neighbors=k, p=p)
        knn.fit(X_train, y_train)

        train_pred = knn.predict(X_train)
        test_pred = knn.predict(X_test)

        train_acc = accuracy_score(y_train,train_pred)
        test_acc = accuracy_score(y_test,test_pred)

        train_err[k].append(1-train_acc)
        test_err[k].append(1-test_acc)

train_opt["(d)i.B."] = min([min(train_err[6]), min(train_err[11]), min(train_err[26])])        
for k in K:
    print(k)
    print(pd.DataFrame(list(zip(P, test_err[k])), columns = ['log10P','Test Error']))
```

##### The best log10(p) is 0.6 when k = 6 with lowest test error 0.06

```python
#C. which becomes Chebyshev Distance with p  
K = np.arange(1, 197, 5)
train_err = []
test_err = []

for k in K:
    knn = KNeighborsClassifier(n_neighbors=k, p=math.inf) # Chebyshev Distance with p  
    knn.fit(X_train, y_train)
    
    train_pred = knn.predict(X_train)
    test_pred = knn.predict(X_test)
    
    train_acc = accuracy_score(y_train,train_pred)
    test_acc = accuracy_score(y_test,test_pred)
    
    train_err.append(1-train_acc)
    test_err.append(1-test_acc)

train_opt["(d)i.C."] = min(train_err[1:]) 
pd.DataFrame(list(zip(K,test_err)), columns = ['K', 'Test Error'])
```

###### when k = 16 for Chebyshev Distance has lowest test_error 0.08

```python
print(min(test_err), K[np.argmin(test_err)])
```

#### ii. Mahalanobis Distance.

```python
train_err = []
test_err = []

for k in K:
    knn = KNeighborsClassifier(n_neighbors=k, algorithm='auto', metric='mahalanobis',metric_params={'V': np.cov(X_train, rowvar=False)})
    knn.fit(X_train, y_train)
    
    train_pred = knn.predict(X_train)
    test_pred = knn.predict(X_test)
    
    train_acc = accuracy_score(y_train,train_pred)
    test_acc = accuracy_score(y_test,test_pred)
    
    train_err.append(1-train_acc)
    test_err.append(1-test_acc)

train_opt["(d)ii."] = min(train_err[1:]) 
pd.DataFrame(list(zip(K,test_err)), columns = ['K', 'TestError'])
```
